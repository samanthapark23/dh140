```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
hamlet = document.text
```

```python
for i in ['\n','[',']','','','']:
    hamlet = hamlet.replace(i,' ')
```

```python
print (hamlet)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sentences = sent_tokenize(hamlet)
sentences
```

```python
words = word_tokenize(hamlet.lower())
```

```python
words
```

```python
extrastop = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve","i.","'d", "'t"]
```

```python
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
myStopWords
```

```python
wordsNoStopWords = [w for w in words if w not in myStopWords]
wordsNoStopWords
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStopWords)
```

```python
freq
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

The top 20 frequent words gives a good sense of the important characters in the play as the list includes numerous character names such as Hamlet, Horatio, Claudius, Polonius, and Gertrude. However, the list doesn't really give us a lot of clues regarding the plot.

```python
listOfNewsWords[0]
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfNewsWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords[0]
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=2, num_words=20):
    print(i)
```

The model chose numerous people's names for the first topic such as "Melissa", "Statham", "Jason", and "McCarthi". It also chose words that seemed to convey a lighthearted and funny mood such as "hilarious", "fresh", "comedi", and "funny". For the second topic, the model chose a lot of words that are complimentary such as "great", "best", "amaz", and "good". It also chose words that allude to the art of the movie such as "screenplay" and "edit". Based on these word choices, I think the model did a pretty good job at splitting up the two different topics of reviews of comedy and drama.
