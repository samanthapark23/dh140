#### Tokenize the words, remove stopwords, stem or lemmatize the words, and calculate the word frequencies

```python
import nltk
```

```python
import requests
```

```python
response = requests.get('http://shakespeare.mit.edu/hamlet/full.html')
```

```python
response
```

```python
from bs4 import BeautifulSoup
```

```python
html_string = response.text
```

```python
document = BeautifulSoup(html_string, "html.parser")
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
hamlet = document.text
```

```python
for i in ['\n','[',']','’','”','“']:
    hamlet = hamlet.replace(i,' ')
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
words = word_tokenize(hamlet.lower())
```

```python
words
```

```python
extrastop = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve","i.","'d", "'t"]
```

```python
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
```

```python
wordsNoStopWords = [w for w in words if w not in myStopWords]
wordsNoStopWords
```

```python
[w for w in word_tokenize(wordsNoStopWords[0].lower()) if w not in myStopWords]
```

```python
listOfNewsWords = []
for i in wordsNoStopWords:
    listOfNewsWords.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
listOfNewsWords[0]
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfNewsWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStopWords)
```

```python
freq
```

```python
type(freq)
```

```python
import pandas as pd
```

```python
hamletdf = pd.DataFrame(list(freq.items()), columns = ["Word","Frequency"]) 
hamletdf
```

#### Make a scatter plot of the target values vs this feature column's values and comment on how the plotted points match up with the correlation coefficient

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
features_df.plot(kind='scatter',
        x='bmi',
        y='target');
```

The plotted points match up with the correlation coefficient of 0.59 as they show a somewhat strong positive trend. 

#### Using this feature and target, perform linear regression with sklearn's LinearRegression. Print the coefficients of the model. Plot the linear fit on top of the scatter plot. Calculate (or output) the mean squared error and R-squared values for your fit


```python
X = features_df[['bmi']]
y = features_df['target']
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
plt.plot(X,y,'ro',X,ytrain,'b-');
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```
